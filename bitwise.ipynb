bitwise operator

1.Bitwise and operator($)
if both bits are 1 then result is 1

 2.Bitwise or operator (|)
 3. Bitwise xor operator (^)
 4.Bitwise not operator (~)

```python
# AND operator
x=10
y=8
z= x & y
print(z)
print(bin(13))
```

```python
# OR operator
x=13
y=11
z= x | y
print(z)
```

```python
# XOR operator
x=10
y=12
z= x ^ y
print(z)
```

```python
# NOT operator
x=10
z= ~x
print(z)
```

```python
# left shift
x=10
y=2
z= x<<y
print(z)
```

```python
# right shift
x=10
y=6
z= x>>y
print(z)
```
