the date time object has a method  for formating date object into readable string
tech method is called strftime() 

```python
import datetime
x= datetime.datetime(2015,11,28)
print(x.strftime("%d/%m/%y"))
print(x.strftime("%d-%m-%y"))
print(x.strftime("%a")) #day
print(x.strftime("%A")) #day
print(x.strftime("%w")) #day
print(x.strftime("%B")) #month
print(x.strftime("%I")) #hour
print(x.strftime("%M")) #minute
print(x.strftime("%S")) #second
```
