function:- 

```python
def functionname():
    print("no argument and no return type")#no argument and no return type

functionname()
```

```python
def functionname(x):
    print("with argument and no return type",x) #with argument and no return type

functionname(12)
```

```python
def add(a, b):
    return a + b

print(add(4, 2)) #with argument and return type
```

```python
def return_without_argument():
    x=10
    return x

y= return_without_argument()
print(y)
```

```python
def return_without_argument(x):
    return x

y= return_without_argument(100)
print(y)
```

local variale and global variable



```python
def func():
    x = 10    # local variable
    print("Local x =", x)

func()
print(x)  
```

```python
x = 20    # global variable

def func():
    print("Inside function:", x)

func()
print("Outside function:", x)
```
