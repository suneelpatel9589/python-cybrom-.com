```python
#filter
def even(x):
    if x%2==0:
        return True
    else:
        return False
ls=[0,5,6,8,9,12,13]
ans=list(filter(even,ls)) #filter(lambda x:x%2==0,l)
print(ans)
```

```python
#map

ls=[0,5,6,8,9,12,13]
ans=list(map(lambda x:x*3,ls)) 
print(ans)
```

```python

from functools import *

ls=[0,5,6,8,9,12,13]
ans=reduce(lambda x,y:x+y,ls) 
print(ans)
```
