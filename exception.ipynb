FILE HANDLING IN PYTHON:-
-------------------------
File handling is an important part of any web scripting language.
Python has capability to handle a file.
Using python we can create a new file as well as we can read the file.
To handle a file first we usen open() method.
e.g:-
-----
f=open("krish.txt")
open():-This method is used to open a file in specified mode.This method has two parameters.First is the name
of the file and second is the mode to open.
Syntax:
---------
open(filename or path,mode)

mode:-
------
read(r):-Read only mode.This mode is used to read the file if file not present then this mode generate an error.
Write(w):-Write mode,This mode is used to write data into file.If file does not present then this mode create a new file.If file exists then replace the file.
Append mode(a):Appenfd mode,This mode append the data at the end of file
