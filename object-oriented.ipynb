Inheritance is a feature of Object-Oriented Programming that allows one class to acquire (or inherit) the properties and behaviors (methods and variables) of another class.

type of Inheritance

```python
# single inheritance

# Parent Class
class College:
    def college_name(self):
        print("ABC Engineering College")

# Child Class
class Student(College):
    def student_info(self):
        print("‡§®‡§æ‡§Æ: ‡§∞‡§æ‡§π‡•Å‡§≤ ‡§∂‡§∞‡•ç‡§Æ‡§æ, ‡§∂‡§æ‡§ñ‡§æ: ‡§ï‡§Ç‡§™‡•ç‡§Ø‡•Ç‡§ü‡§∞ ‡§∏‡§æ‡§á‡§Ç‡§∏")

# ‡§ë‡§¨‡•ç‡§ú‡•á‡§ï‡•ç‡§ü ‡§¨‡§®‡§æ‡§®‡§æ
s = Student()
s.college_name()   # Parent class ‡§ï‡§æ method
s.student_info()   # Child class ‡§ï‡§æ method
```

Hierarchical Inheritance:-

Hierarchical inheritance means one parent (base) class is inherited by multiple child classes.

```python
# Parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class 1
class Dog(Animal):
    def speak(self):
        print("Dog barks")

# Child class 2
class Cat(Animal):
    def speak(self):
        print("Cat meows")


# Creating objects
dog = Dog()
cat = Cat()

dog.speak()   # Output: Dog barks
cat.speak()   # Output: Cat meows
```

```python
class Parent:
    def __init__(self):
        print('this is constructor ')
    def parentfunc(self):
        print('this is a parent function')
class base1(parent):
    def  __init__(self):
        parent.__init__(self)
        print('this is a base1 class consstructor')
    def basefunc1(self):
        print('this is a base1 function')
class base2(parent):
    def  __init__(self):
        super().__init__()
        print('this is a base2 class consstructor')
    def basefunc1(self):
        print('this is a base2 function')

ob=base1()
ob=base2()
    
```

Hybrid Inheritance:

Hybrid Inheritance = ‡§¶‡•ã ‡§Ø‡§æ ‡§¶‡•ã ‡§∏‡•á ‡§ú‡•ç‡§Ø‡§æ‡§¶‡§æ inheritance types ‡§ï‡§æ combination‡•§
Single, Multiple, Multilevel, Hierarchical inheritance

```python
# Parent class
class A:
    def show(self):
        print("Class A")

# Child class 1
class B(A):
    def show(self):
        print("Class B")

# Child class 2
class C(A):
    def show(self):
        print("Class C")

# Hybrid class
class D(B, C):
    def show(self):
        super().show()   # Parent classes ‡§Æ‡•á‡§Ç ‡§∏‡•á first method call
        print("Class D")

obj = D()
obj.show()
```
