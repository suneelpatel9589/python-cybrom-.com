```python
class Person:
    def __init__(self, name, age):  # Constructor
        self.name = name           # Instance variable
        self.age = age             # Instance variable

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old")

# Object ‡§¨‡§®‡§æ‡§®‡§æ
p1 = Person("Alice", 25)
p1.greet()
```

```python
# instance variable:- 

class MyClass:
    def __init__(self, name):
        self.name = name  # instance variable

obj1 = MyClass("Alice")
obj2 = MyClass("Bob")

print(obj1.name)  # Alice
print(obj2.name)  # Bob
```

```python
# class variable:- 

class MyClass:
    school = "XYZ School"  # class variable

obj1 = MyClass()
obj2 = MyClass()

print(obj1.school)  # XYZ School
print(obj2.school)  # XYZ School
```

```python
class mycls:
    def __init__(self):
        print("outer class constructor")
    def outerclsfunc(self):
        print('user define outer function')
    class inner:
        def __init__(self):
            print("this is a inner constructor")
        def innerclsfunc(self):
            print('this is inner class user define function')
ob=mycls()
ob.outerclsfunc()
ib=ob.inner()
ib.innerclsfunc()
            
```

first piller in oops:- Inheritance
it means reuselibity 
Inheritance is a feature of Object-Oriented Programming that allows one class to acquire (or inherit) the properties and behaviors (methods and variables) of another class.

type of Inheritance

1.single line Inheritance
2. Multiple Inheritance
3. Multilevel Inheritance
4. Hierarchical Inheritance
5. Hybrid Inheritance

```python
# single inheritance

# Parent Class
class College:
    def college_name(self):
        print("ABC Engineering College")

# Child Class
class Student(College):
    def student_info(self):
        print("‡§®‡§æ‡§Æ: ‡§∞‡§æ‡§π‡•Å‡§≤ ‡§∂‡§∞‡•ç‡§Æ‡§æ, ‡§∂‡§æ‡§ñ‡§æ: ‡§ï‡§Ç‡§™‡•ç‡§Ø‡•Ç‡§ü‡§∞ ‡§∏‡§æ‡§á‡§Ç‡§∏")

# ‡§ë‡§¨‡•ç‡§ú‡•á‡§ï‡•ç‡§ü ‡§¨‡§®‡§æ‡§®‡§æ
s = Student()
s.college_name()   # Parent class ‡§ï‡§æ method
s.student_info()   # Child class ‡§ï‡§æ method
```
