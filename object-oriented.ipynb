```python
class Person:
    def __init__(self, name, age):  # Constructor
        self.name = name           # Instance variable
        self.age = age             # Instance variable

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old")

# Object ‡§¨‡§®‡§æ‡§®‡§æ
p1 = Person("Alice", 25)
p1.greet()
```

```python
# instance variable:- 

class MyClass:
    def __init__(self, name):
        self.name = name  # instance variable

obj1 = MyClass("Alice")
obj2 = MyClass("Bob")

print(obj1.name)  # Alice
print(obj2.name)  # Bob
```

```python
# class variable:- 

class MyClass:
    school = "XYZ School"  # class variable

obj1 = MyClass()
obj2 = MyClass()

print(obj1.school)  # XYZ School
print(obj2.school)  # XYZ School
```

```python
# local variable

class MyClass:
    def greet(self):
        msg = "Hello World"  # local variable
        print(msg)

obj = MyClass()
obj.greet()  # Hello World
#  print(msg) -> error, ‡§ï‡•ç‡§Ø‡•ã‡§Ç‡§ï‡§ø msg ‡§∏‡§ø‡§∞‡•ç‡§´ method ‡§Æ‡•á‡§Ç exist ‡§ï‡§∞‡§§‡§æ ‡§π‡•à
```
