```python
def isprime(x):    #default argument
    cnt=0
    for i in range(1,x+1):
        if(x%i==0):
            cnt+=1
    if(cnt==2):
        print('prime')
    elif(cnt==3):
        print('three factor')
    elif(cnt==4):
        print('four factor')
    else:
        print('more thane four factor')
        
isprime(6)
```

```python
def arm():
    n=int(input('enter a number'))
    m=n
    ans=0
    p=len(str(n))
    while(n>0):
        r=n%10
        ans=ans+r**p
        n=n//10
    if(ans==m):
        print('armstrong')
    else:
        print('not armstrong')

arm()
```

Recursion: Jab ek function khud ko call karta hai.
Har recursive function ke do main parts hote hain:

1.base condition:- Ye wo condition hoti hai jisme function khud ko dubara call karna band kar deta hai.
Iske bina recursion kabhi rukega nahi → infinite recursion ho jayega.

2.recursion:-Yaha function khud ko call karta hai aur problem ko chhote part me todta hai.
Har call original problem ko thoda chhota karta hai.

note;- - Jab ye condition true hoti hai, function aage recursive call nahi karega.
   - Yahi recursion ko rokne ka signal hota hai

```python
def fact(n):
    if(n==0): #base condition
        return 1
    ans=n*fact(n-1) #recursion relation
    return ans
fact(5)

```

```python
def fact(n):
    if(n==0): #base condition
        return 0
    return n+fact(n-1) #recursion relation
x=fact(5)
print(x)
```
