File handling in python-> is is an important part of any web scripting language.
                          python has capability to handle a file.
                          using python we can create a new file asa well as we can reade the file.
                          to handle a file we use open() method.
                          e.g.->  f=open("krish.txt")
open method-> this method is use to open a file in specified mode . this method has two parameters
              1. name of the file
              2. mode to open
              syntax-> open (Filename or path, mode)

mode -> 1. read(r) -> read only mode. this mode is used to read the file if file not present then this mode generate an error.
        2. write(w) -> write mode. this mode is used to write data into file. If file does not present then this mode create a new file. If file exist 
                                   replace that file
        3. append(a) -> Append mode. this mode append the data at the end of file.
        4. extend(x) -> extend mode. this mode use to write data into file if file does note exist this mode create a new file. 
                                      if file exist this mode will not replace the old file.
method -> 1. read method -> this method is use to read the data of the file from the file object.
          2. read line method -> this method is used to read data inside the file one line at a time. 
          3. write method -> this method is used to write data into file.

```python
# write

f=open("krish1.txt","w")
f.write("This is krish file")
print("File created successfully")
```

```python
# extend

f=open("krish.txt","x")
f.write("This is krish file...!!!")
print("File created successfully")
```

```python
# read

f=open("krish1.txt","r")
x=f.read()
print("File created successfully",x)
```

```python
# read , readline

f=open("krish1.txt","r")
x=f.readline()
print("File created successfully",x)
```
