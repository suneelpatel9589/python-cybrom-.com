üëâ operators (+ - * /) ‡§ï‡•ã ‡§Ö‡§™‡§®‡•á class ‡§ï‡•á ‡§≤‡§ø‡§è ‡§®‡§Ø‡§æ meaning ‡§¶‡•á‡§®‡§æ

```python
class mycls:
    def __init__(self, var1):
        self.var1 = var1

    def __mul__(self, other):
        return self.var1 * other.var1

ob1 = mycls(100)
ob2 = mycls(200)

print(ob1 * ob2)
```

Common Operator Overloading Methods
Operator	Method
+       	__add__
-	        __sub__
*       	__mul__
/       	__truediv__
//q     	__floordiv__
%       	__mod__
==	        __eq__
<	        __lt__

```python
class mycls:
    def __init__(self, var1):
        self.var1 = var1

    def __sub__(self, other):
        return self.var1- other.var1

ob1 = mycls(100)
ob2 = mycls(200)

print(ob1 - ob2)
```

```python
class mycls:
    def __init__(self, var1):
        self.var1 = var1

    def __add__(self, other):
        return self.var1+other.var1

ob1 = mycls(100)
ob2 = mycls(200)

print(ob1 + ob2)
```

```python
class mycls:
    def __init__(self, var1):
        self.var1 = var1

    def __truediv__(self, other):
        return self.var1/other.var1

ob1 = mycls(12)
ob2 = mycls(2)

print(ob1 / ob2)
```
